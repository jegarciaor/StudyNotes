# Capítulo 2

Se comienza analizando el algoritmo `insertion-sort` con el cuál se pretende mostrar el uso de los **pseudocódigos**, como también analizar su implementación para solucionar algo más general **SORTING PROBLEM** y cuya definición tiene la siguiente forma:



**Entrada:** Una secuencia de $n$ números tipo $(a_1, a_2, \ldots, a_n)$.


**Salida:** Una secuencia $(a_1^`, a_2^`, a_3^`, \ldots, a_n^`)$ donde se cumple que $a_1^` \leq a_2^` \leq a_3^` \leq \ldots \leq a_n^`$.


A continuación se presenta el pseudocódigo para analizar el algoritmo. 

`Insertion-Sort(A)`

`1. for j = 2 to A.length:`

`2.     key = A[j]`

`3.     i = j - 1`

`4.     while i>0 and A[i]> key:`

`5.         A[i+1] = A[i]`

`6.         i = i-1`

`7.     A[i+1] = key`

**Observaciones:** 

1. Nótese que en la línea $1$ se inicia el ciclo for y, a partir de ahí, se empieza a orgarnizar la parte de la cadena que está antes de este valor.

    
2. **key** es una variable auxiliar que va a permitir realizar comparaciones.


3. Desde la línea $4$ hasta $6$ hay un ciclo while encargado de ordenar los valores anteriores al valor $A[j]$.


### Implementación del algoritmo

Se procede a implementar el algoritmo haciendo uso de Python3.


```python
# Implementación Insertion_sort.

def InsertionSort(secuencia):
    for j in range(1, len(secuencia)):
        key = secuencia[j]
        i = j - 1
        
        while (i > -1) and (secuencia[i] > key):
            secuencia[i + 1] = secuencia[i]
            i = i - 1
        
        secuencia[i + 1] = key
    
    return secuencia
```

```python
# Creamos una lista para ponerla en orden mediante el algoritmo.
lista = [5, 2, 4, 6, 1, 3]
```

```python
# Ensayamos la lista creada en el algoritmo.
InsertionSort(lista)
```
