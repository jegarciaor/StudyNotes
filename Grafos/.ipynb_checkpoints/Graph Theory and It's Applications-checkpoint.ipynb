**Observacin:** Del teorema $1.1.4$ podemos saber cundo una sucesin de numeros determinan un grafo no simple; sin embago, queda la pregunta: cules son las condiciones necesarias para determinar un grafo simple?

La pregunta anterior lleva a lo siguiente:


Dada una sucesin $(d_1, d_2, \cdots, d_n)$ de enteros positivos, y no creciente, cmo hacemos para determinar si los valores de esta sucesin determinan un grafo simple?

La respuesta a este problema la dan Havel(1955) y Hakimi(1961) con la implementacin de un algoritmo recursivo que tiene la siguiente estructura:


**Input:** Sucesin $(d_1, d_2, \cdots, d_n)$ decreciente de nmeros positivos.

**Output:** True o False indicando si la sucesin representa o no un grafo simple.


El pseudocdigo y teora necesaria para llegar a la implementacin del algoritmo se encuentran en las pginas $9$ y $10$ del libro.

```python
# Implementacin del algoritmo de Havel-Hakimi.

def havelHakimi(secuencia):
    secuencia.sort(reverse=True)
    if secuencia[0] >= len(secuencia) or secuencia[-1] < 0:
        return False
    elif secuencia[0] == 0:
        return True
    else:
        valor_quitado = secuencia.pop(0)
        for i in range(valor_quitado):
            secuencia[i] -= 1
        secuencia.sort(reverse=True)

    return havelHakimi(secuencia)
```
